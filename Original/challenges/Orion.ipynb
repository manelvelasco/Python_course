FFT Challenge
========
Let's put some noise and remove some noise from this fantastic image (acquired by Velasco and Perera, by the way). 


```python
import numpy as np
from scipy import fftpack,ndimage,outer

import matplotlib.pyplot as plt
img = ndimage.imread('orionnebula.jpg',flatten=1)
plt.imshow(img,cmap=plt.cm.gray)


```

```python
x = np.arange(0,shape(img)[1])
y = np.arange(0,shape(img)[0])
xped = 50
yped = 50
sigx = sin(2 * np.pi * x / xped)
sigy = cos(2 * np.pi * y / yped)
img2 = outer( sigy, sigx )

plt.imshow(img2,cmap=plt.cm.gray)
print x
```

```python
from scipy import misc
img3 = img/2 + 100*img2/2 

plt.imshow(img3,cmap=plt.cm.gray)
misc.imsave('orionnebulaN.jpg',img3)

```

Now, we have the problem image with some noise. Then, how do we remove the noise?

First import the image with plt.imread



```python
I = plt.imread('orionnebulaN.jpg')
fftI=fftpack.fft2(I)     
plt.imshow(log(abs(fftI)),cmap=plt.cm.gray)
```

```python
abs(fftI[1:2,1:2])
psd2D = np.abs(  fftpack.fftshift( fftI ) )**2
plt.imshow(log(psd2D),cmap=plt.cm.gray)
```

Finally , we will apply a brick-wall frequency-domain filter to the data, zeroing out all FFT outputs that correspond to a frequency greater than 0.005 Hz, then inverse-transforming to get a time-domain signal again

```python
N=20
sfftI=np.copy(fftI)
sfftI[N:(shape(sfftI)[0]-N),N:(shape(sfftI)[1]-N)]=0
fI=fftpack.ifft2(sfftI)

plt.imshow(abs(fI),cmap=plt.cm.gray)
```
