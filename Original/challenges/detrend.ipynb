Signal processing challenge I
===================

Use scipy.signal to: 

* Create a linear signal.
* Add a random noise process.
* Detrend the signal. 

```python
t = np.linspace(0, 5, 100)
x = t + np.random.normal(size=100)
```

```python
from matplotlib import pyplot as pl
from scipy.signal import detrend
pl.plot(t, x, linewidth=3)
pl.plot(t, detrend(x), linewidth=3)
```
