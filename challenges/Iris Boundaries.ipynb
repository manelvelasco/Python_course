# K-nearest neighbours vs. SVC Challenge

This challenge invites you to develope the following:

* Get all variables from Iris dataset.
* Compute a 2D-PCA model of the Iris dataset.
* Compute and plot the decision boundaries for a k-NN and SVC classifier of your choice. 

First let's import the tools that we need.

```python
import numpy as np
import pylab as pl
from sklearn import svm, neighbors, datasets, decomposition
```

## Get All variables from the Iris dataset

```python
iris = datasets.load_iris()
X, y = iris.data, iris.target
```

## Compute a 2D-PCA model of the Iris dataset.

```python
pca = decomposition.PCA(n_components=2)
pca.fit(X)
T = pca.transform(X)
```

## Compute and plot the decision boundaries for a k-NN and SVC classifier of your choice.

```python
h=0.03
x_min, x_max = T[:, 0].min() - 1, T[:, 0].max() + 1
y_min, y_max = T[:, 1].min() - 1, T[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
```

Let's define our classifiers!

```python
C, gamma = 1.0 , .7 # SVM regularization parameter
k=5 # k parameter for k-NN
svcl = svm.SVC(kernel='linear', C=C).fit(T, y)
svcr = svm.SVC(kernel='rbf', gamma=gamma, C=C).fit(T, y)
knn = neighbors.KNeighborsClassifier(5).fit(T,y)

titles = ['SCV Lineal',
          'SVC Radial',
          '5-NN']
```

Hey, we can iterate through a list of models!

```python

for i, classifier in enumerate((svcl, svcr, knn)):
    #classifier will take each one of the models
    
    pl.subplot(2, 2, i + 1) # I suspect the 1 indexing is for matlabers
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    pl.contourf(xx, yy, Z, cmap=pl.cm.Paired)
    pl.axis('off')

    # Plot also the training points
    pl.scatter(T[:, 0], T[:, 1], c=y, cmap=pl.cm.Paired)

    pl.title(titles[i])

pl.show()
```
